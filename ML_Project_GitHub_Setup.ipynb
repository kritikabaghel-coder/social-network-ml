## Section 1: Set Up Project Directory Structure

A well-organized directory structure is essential for GitHub projects. Here's what a professional ML project looks like:

````python
import os
from pathlib import Path

# Define the project structure
project_structure = {
    "data": {
        "raw": "Original datasets",
        "processed": "Cleaned and processed data",
    },
    "notebooks": {
        "_": "Jupyter notebooks for analysis"
    },
    "scripts": {
        "_": "Reusable Python scripts"
    },
    "results": {
        "models": "Trained ML models",
        "metrics": "Model evaluation metrics",
        "plots": "Generated visualizations",
    },
    "outputs": {
        "_": "Final project outputs and reports"
    },
    "tests": {
        "_": "Unit tests"
    },
}

# Create the directory structure
base_path = Path(".")

print("📁 Project Directory Structure:\n")
print("project/")
for folder, subfolders in project_structure.items():
    print(f"├── {folder}/")
    if isinstance(subfolders, dict):
        for subfolder, description in subfolders.items():
            if subfolder != "_":
                print(f"│   ├── {subfolder}/")
            print(f"│   │   # {description}")
    print()

# Create directories
for folder in project_structure.keys():
    os.makedirs(folder, exist_ok=True)
    if isinstance(project_structure[folder], dict):
        for subfolder in project_structure[folder].keys():
            if subfolder != "_":
                os.makedirs(f"{folder}/{subfolder}", exist_ok=True)

print("✅ Directory structure created successfully!")
````

## Section 5: Configure Output Directories

Automatically organize and save outputs with proper structure:

````python
from pathlib import Path
from datetime import datetime
import json

print("📁 OUTPUT MANAGEMENT SYSTEM\n")

# Create output directory manager
class OutputManager:
    def __init__(self, base_dir="outputs"):
        self.base_dir = Path(base_dir)
        self.timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        
        # Create subdirectories
        self.dirs = {
            "models": self.base_dir / "models",
            "metrics": self.base_dir / "metrics",
            "plots": self.base_dir / "plots",
            "reports": self.base_dir / "reports",
            "logs": self.base_dir / "logs",
        }
        
        for dir_path in self.dirs.values():
            dir_path.mkdir(parents=True, exist_ok=True)
    
    def save_model(self, model, name):
        """Save ML model with timestamp."""
        import pickle
        filepath = self.dirs["models"] / f"{name}_{self.timestamp}.pkl"
        with open(filepath, "wb") as f:
            pickle.dump(model, f)
        print(f"✓ Model saved: {filepath}")
        return filepath
    
    def save_metrics(self, metrics, name):
        """Save metrics as JSON."""
        filepath = self.dirs["metrics"] / f"{name}_{self.timestamp}.json"
        with open(filepath, "w") as f:
            json.dump(metrics, f, indent=2)
        print(f"✓ Metrics saved: {filepath}")
        return filepath
    
    def save_plot(self, fig, name):
        """Save matplotlib figure."""
        filepath = self.dirs["plots"] / f"{name}_{self.timestamp}.png"
        fig.savefig(filepath, dpi=300, bbox_inches='tight')
        print(f"✓ Plot saved: {filepath}")
        return filepath
    
    def save_report(self, content, name):
        """Save text report."""
        filepath = self.dirs["reports"] / f"{name}_{self.timestamp}.md"
        with open(filepath, "w") as f:
            f.write(content)
        print(f"✓ Report saved: {filepath}")
        return filepath

# Example usage
print("📋 Example: Using OutputManager\n")

output_manager = OutputManager()

print("Output directories created:")
for key, path in output_manager.dirs.items():
    print(f"  ✓ {key}: {path}")

print("\n📊 Typical Output Structure:")
print("""
outputs/
├── models/
│   ├── model_20260324_132501.pkl
│   └── vectorizer_20260324_132501.pkl
├── metrics/
│   ├── train_metrics_20260324_132501.json
│   └── test_metrics_20260324_132501.json
├── plots/
│   ├── confusion_matrix_20260324_132501.png
│   ├── roc_curve_20260324_132501.png
│   └── feature_importance_20260324_132501.png
├── reports/
│   ├── model_summary_20260324_132501.md
│   └── analysis_report_20260324_132501.md
└── logs/
    └── training_log_20260324_132501.txt
""")

print("✅ Output management system ready!")
````
